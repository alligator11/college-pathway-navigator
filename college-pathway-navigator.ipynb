```python
import pandas as pd
```

```python
df = pd.read_csv("/content/round1_josaa_22.csv")
df.head()
```

```python
df.columns = [col.replace(" ","_") for col in df.columns]
```

```python
df.columns
```

```python
df[df['Opening_Rank'].str.contains('P', regex="True")]
```

```python
#Assuming that P does not hold any significant meaning
df['Opening_Rank']=df['Opening_Rank'].replace("P","", regex=True)
df['Closing_Rank']=df['Closing_Rank'].replace("P","", regex=True)
```

```python
df[df['Opening_Rank'].str.contains('P')]
```

```python
data_types_dict = {'Opening_Rank': int, 'Closing_Rank': int}
df = df.astype(data_types_dict)
df.dtypes
```

```python
seat_types=df['Seat_Type'].unique()
seat_types
```

```python
df['Gender'].unique()
```

```python
rank=int(input("Enter your rank: "))
print("Seat_Types:",seat_types)
seat=input("Enter your seat type (copy from the above list): ")
gender=(input("Enter your gender [1 for gender-neutral and 2 for female-only (including supernumerary)]:"))
if gender=='1':
  gender='Gender-Neutral'
elif gender == '2':
  gender='Female-only (including Supernumerary)'
df_output=df.query('Opening_Rank <= @rank and Closing_Rank >= @rank and Seat_Type == @seat and Gender == @gender')
df_output[["Institute","Academic_Program_Name"]]
```
